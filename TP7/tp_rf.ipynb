```python
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer()
```

```python
import pandas as pd
df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
df['target'] = dataset.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('target', axis=1)
X
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
# create an instance of random forest classifier
rf = RandomForestClassifier()
```

```python
print(rf.get_params())
```

### The parameters of a random forest classifier are the following : 

{'bootstrap': True, 'ccp_alpha': 0.0, 'class_weight': None, 'criterion': 'gini', 'max_depth': None, 'max_features': 'sqrt', 'max_leaf_nodes': None, 'max_samples': None, 'min_impurity_decrease': 0.0, 'min_samples_leaf': 1, 'min_samples_split': 2, 'min_weight_fraction_leaf': 0.0, 'monotonic_cst': None, 'n_estimators': 100, 'n_jobs': None, 'oob_score': False, 'random_state': None, 'verbose': 0, 'warm_start': False}

### Brief explanation :

- **bootstrap**: This parameter determines whether bootstrap samples are used when building trees. Bootstrap sampling means that each tree in the forest is built on a random sample with replacement from the original dataset. It helps introduce randomness into the trees, which can improve the diversity of the ensemble and reduce overfitting. By default, it's set to True, meaning bootstrap samples are used.

- **ccp_alpha**: This parameter is the complexity parameter used for Minimal Cost-Complexity Pruning. It determines the amount of regularization applied to the trees by pruning them based on their cost-complexity. Higher values of ccp_alpha lead to more aggressive pruning, resulting in simpler trees. The default value is 0.0, meaning no pruning based on cost-complexity is performed by default.

- **class_weight**: This parameter is used to assign weights to classes in the input data. It can be set to either "balanced" or a dictionary where the keys are class labels and the values are the corresponding weights. This is useful when dealing with imbalanced datasets, as it helps the model to pay more attention to minority classes. By default, it's set to None, meaning all classes are treated equally.

- **criterion**: This parameter determines the function used to measure the quality of a split. The two supported criteria are "gini" for the Gini impurity and "entropy" for the information gain. Gini impurity measures the probability of incorrectly classifying a randomly chosen element if it were randomly labeled according to the distribution of labels in the node, while entropy measures the level of impurity or disorder in the data. By default, it's set to "gini".

- **max_depth**: This parameter specifies the maximum depth of the trees in the forest. A deeper tree can model more complex relationships in the data but is more likely to overfit. By default, it's set to None, meaning nodes are expanded until all leaves are pure or until they contain less than min_samples_split samples.

- **max_features**: This parameter determines the maximum number of features considered for splitting a node. It can be set to "auto", "sqrt", "log2", or an integer representing the number of features. By default, it's set to "sqrt", meaning the square root of the total number of features is considered at each split.

- **max_leaf_nodes**: This parameter specifies the maximum number of leaf nodes in each tree. It helps control the growth of trees to prevent overfitting. By default, it's set to None, meaning there is no maximum limit on the number of leaf nodes.

- **max_samples**: This parameter specifies the maximum number of samples used to train each tree. It can be set to an integer or None. If None, all samples are used. By default, it's set to None.


```python
rf.fit(X_train, y_train)
```

```python
train_score = rf.score(X_train, y_train)
test_score = rf.score(X_test, y_test)

print("Training Score:", train_score)
print("Testing Score:", test_score)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = rf.predict(X_test)

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

### Choose the best hyperparams for the rf classifier


```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'max_depth': [3, 5, 7],
    'n_estimators': list(range(10, 200, 10))
}


grid_search = GridSearchCV(estimator=rf,
                           param_grid=param_grid,
                           cv=5)
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

### Instanciate the rf classifier using the best hyperparams and fit it on our data

```python
rf_best = RandomForestClassifier(n_estimators=60, max_depth=7, random_state=0)
rf_best.fit(X, y)
```

```python
train_score = rf_best.score(X_train, y_train)
test_score = rf_best.score(X_test, y_test)

print("Training Score:", train_score)
print("Testing Score:", test_score)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = rf_best.predict(X_test)

conf_matrix_ = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
Importance= pd.DataFrame ({'Importance':rf_best.feature_importances_*100},
index=X.columns)
Importance.sort_values (by='Importance', axis=0, ascending=True).plot(kind='barh',
color='r')
plt.xlabel('feature importance')
plt.gca().legend_=None
```

#### Do the most important features change if you choose a different number of trees in a random forest?
The most important features can change if we choose a different number of trees in a random forest. The feature importance values are calculated based on the average impurity decrease from splitting on a feature across all trees in the forest. Changing the number of trees can affect how these impurity decreases are aggregated and thus impact the relative importance of features.
